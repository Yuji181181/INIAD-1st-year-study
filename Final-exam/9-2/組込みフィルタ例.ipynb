| フィルタ                | 説明                                                                                                                                                                            |
|-------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| add                    | 入力値に対して引数の値を加算します。                                                                                                                                             |
| addslashes             | 引用符の前にスラッシュを追加します。CSV などの文字列をエスケープする際に便利です。                                                                                                   |
| capfirst               | 入力値の先頭の文字を大文字に変換します。最初の文字がアルファベットでなければ効果はありません。                                                                                       |
| center                 | 入力値を引数で指定された幅のフィールド内に中央寄せします。                                                                                                                         |
| cut                    | 入力値の中から、引数に指定した値を全て取り除きます。                                                                                                                               |
| date                   | 引数に指定した書式で日付をフォーマットします。                                                                                                                                   |
| default                | 入力の評価値が False の場合、引数に指定したデフォルト値を使います。そうでなければ、入力値を使います。                                                                                |
| default_if_none        | 入力値が None の場合、引数に指定したデフォルト値を使います。そうでなければ、入力値を使います。                                                                                       |
| dictsort               | 辞書のリストを入力として、引数に指定したキーでリストをソートして返します。                                                                                                         |
| dictsortreversed       | 辞書のリストを入力に取り、引数に指定したキーでリストを逆順にソートして返します。                                                                                                     |
| divisibleby            | 値が引数の値で割り切れる場合に True を返します。                                                                                                                                 |
| escape                 | 入力文字中にある HTML の特殊文字をエスケープします。                                                                                                                               |
| escapejs               | JavaScript の文字列リテラルとして扱うために文字エスケープを行います。                                                                                                              |
| filesizeformat         | 入力値を人間が読みやすいファイルサイズ表現（例：『13 KB』, 『4.1 MB』）に変換します。                                                                                              |
| first                  | リスト中の最初の要素を返します。                                                                                                                                                  |
| floatformat            | 浮動小数点数を小数点以下1桁に丸めます。ただし小数部分がない時には整数部分だけを表示します。                                                                                          |
| force_escape           | 文字列に HTML エスケープを適用します。                                                                                                                                            |
| get_digit              | 整数の入力値から、引数で指定された桁の数字を返します。                                                                                                                             |
| iriencode              | IRI（国際化リソース識別子）を URL に適した文字列に変換します。                                                                                                                      |
| join                   | リストを指定の文字列でつなぎます。                                                                                                                                               |
| last                   | リストの末尾の要素を返します。                                                                                                                                                    |
| length                 | 入力値の長さを返します。                                                                                                                                                         |
| length_is              | 入力値の長さと引数が等しければ True を返します。                                                                                                                                 |
| linebreaks             | プレーンテキストの改行を HTML タグに変換します。                                                                                                                                 |
| linebreaksbr           | プレーンテキストの改行を `<br />` に変換します。                                                                                                                                 |
| linenumbers            | テキストを行番号付きで表示します。                                                                                                                                                |
| ljust                  | 入力値を指定した幅のフィールド内で左詰めします。                                                                                                                                   |
| lower                  | 文字列をすべて小文字に変換します。                                                                                                                                               |
| make_list              | 入力値をリストに変換します。                                                                                                                                                     |
| phone2numeric          | 電話番号を数値だけの番号に変換します。                                                                                                                                            |
| pluralize              | 入力値が 1 でない場合に、複数形を表す接尾辞を返します。                                                                                                                             |
| pprint                 | `pprint.pprint()` のラッパーです。                                                                                                                                               |
| random                 | リストからランダムな要素を返します。                                                                                                                                              |
| rjust                  | 入力値を指定した幅のフィールド内で右詰めします。                                                                                                                                   |
| safe                   | 文字列を HTML エスケープせずに出力します。                                                                                                                                         |
| safeseq                | シーケンスの各要素に対して `safe` フィルタを適用します。                                                                                                                           |
| slice                  | リストに対するスライスを返します。                                                                                                                                               |
| slugify                | 文字列をスラッグ化します（スペースをハイフンに変換し、特殊文字を削除します）。                                                                                                        |
| stringformat           | 変数を指定した書式でフォーマットします。                                                                                                                                          |
| striptags              | HTML タグをすべて取り除きます。                                                                                                                                                   |
| time                   | 時刻を指定の書式でフォーマットします。                                                                                                                                            |
| timesince              | 日付を経過時間の形式にフォーマットします。                                                                                                                                        |
| timeuntil              | 指定の日付または時刻までの残り時間を計算します。                                                                                                                                  |
| title                  | 文字列をタイトルケースに変換します。                                                                                                                                             |
| truncatechars          | 文字列を指定した文字数以内に切り詰めます。                                                                                                                                         |
| truncatechars_html     | HTML タグを認識しつつ文字列を指定した文字数以内に切り詰めます。                                                                                                                      |
| truncatewords          | 文字列を指定した単語数以内に切り詰めます。                                                                                                                                         |
| truncatewords_html     | HTML タグを認識しつつ文字列を指定した単語数以内に切り詰めます。                                                                                                                      |
| unordered_list         | 入れ子のリストを HTML の順不同リストに変換します（外側の `<ul>` タグは表示しません）。                                                                                              |
| upper                  | 文字列をすべて大文字に変換します。                                                                                                                                               |
| urlencode              | 入力値を URL エンコーディングします。                                                                                                                                            |
| urlize                 | テキスト内の URL と Email アドレスをクリック可能なリンクに変換します。                                                                                                               |
| urlizetrunc            | `urlize` と同様に動作しますが、指定した文字数で表示を切り詰めます。                                                                                                                   |
| wordcount              | 入力値の単語数を返します。                                                                                                                                                       |
| wordwrap               | 指定した行幅でワードラップします。                                                                                                                                                |
| yesno                  | 入力値（True, False, None）に対応する文字列を返します。                                                                                                                             |
