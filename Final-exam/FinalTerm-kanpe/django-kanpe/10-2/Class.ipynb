# クラス

## 復習（あんま見る必要ないっす）
#### クラスとは
- データを抽象化したもの
- インスタンスの設計図
- データ型と同じ意味
#### インスタンスとは
- データの具体例
- あるクラスの具体的な個体
- データの値と同じ意味
#### クラスの書き方
- クラスは「class 任意のクラス名」と書いて定義
    - クラス名の一文字目は大文字
- クラス変数
    - 共通する性質
- 属性
    - 各インスタンス固有の性質
- メソッド
    - 各インスタンスに共通する操作

![クラスのイメージ](images\クラスの例.png)

#### クラスの継承
- pythonのクラスは別のクラスを継承することができる
    - 親クラスをスーパークラス、子クラスをサブクラスという
- 子クラスのインスタンスは親クラスの属性やメソッドを利用できる！！！！！！！！
- 継承する場合は、クラスの定義でクラス名の後ろに（親クラス）とかく

![クラス継承のイメージ](images\クラスの継承：例.png)

#### 親クラスと子クラスの関係
- 親クラスは、子クラスを汎化したもの
    - より抽象化したもの
- 子クラスは、親クラスを特化したもの
    - より具体的にしたもの

#### 継承の機能
- 子クラスのインスタンスは、親クラスのメソッドも利用できる
