# データベースとは
データを保存・管理するためのシステム全般を指す。  
データの保存形式や構造はシステムごとに異なる。

## 関係データベース
- データをテーブル形式で管理し、テーブル間の関係を明確に管理するデータベース。
- 関係データベースを操作するための標準言語は **SQL**。

# Djangoにおけるデータベース操作
Djangoでは **モデル** を通してデータベースにアクセスすることができる。

![Djangoモデルの仕組み](images/Djangoのモデルの仕組み.png)

## モデル作成の基本
- **Modelクラス** は `django.db.models.Model` クラスを継承して作成する。
- フィールドは `XxxField` のインスタンスとして定義する。
  - **Field** は用いるデータ型に応じて使い分ける必要がある。
  - 型一覧: [Django公式ドキュメント](https://docs.djangoproject.com/ja/4.1/ref/models/fields/#model-field-types)

### 主キー (Primary Key)
- 特に指定がない場合、自動的に追加される `id` フィールドが主キーとして使用される。

# よく用いられるフィールド型

#### CharField
- **カラムの型**: `varchar`  
  ※SQLite3では内部的に `text` 型と同じ処理を受ける（他のデータベースとは異なる点に注意）
- **用途**: ユーザー名やタイトル名など短い文字列を格納する。  
  `max_length=xxx` で最大文字数を指定する必要がある。



#### TextField
- **カラムの型**: `text`  
  ※文字数に制限はない（厳密には存在するが、実用上問題のない範囲）
- **用途**: ブログの本文や自由記述などの長い文字列を格納する。  
  最大文字数を指定する必要がない。



#### IntegerField
- **カラムの型**: `Integer`
- **用途**: いいねの数や年齢など、整数値を格納する。



#### FloatField
- **カラムの型**: `Float`
- **用途**: スコアなどの浮動小数点数を格納する。



#### DateTimeField
- **カラムの型**: `datetime`
- **用途**: 投稿日時などの日付日時を格納する。



#### EmailField
- **カラムの型**: `email`
- **用途**: メールアドレスを格納する。



#### JSONField
- **カラムの型**: `text` （SQLiteの場合）
- **用途**: リストや辞書を格納する。



#### ForeignKey
- **カラムの型**: `Integer`
- **用途**: 他テーブルとの関連を作るときに用いる。
