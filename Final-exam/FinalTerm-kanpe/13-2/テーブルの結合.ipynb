countとgroup byとhavingについて

COUNT(*)は、SQLで使用される集計関数の一つで、指定した条件に一致する行の数をカウントします。たとえば、`SELECT COUNT(*) FROM table_name;`と記述すると、そのテーブル内のすべての行の数が返されます。
具体的には：
- COUNT(id)は、`id`列に値がある行の数をカウントします。
- COUNT(*)は、NULL値を含むすべての行をカウントします。

SELECT 任意の属性, COUNT(任意の属性) FROM あるtable GROUP BY 任意の属性というクエリは、指定した属性を主キーとして、各キーごとに同じ値を持つレコードの数を計算し、表示するものです。具体的には、`GROUP BY`句によって指定した属性でグループ化され、`COUNT`関数によって各グループに含まれるレコードの数が集計されます。

(レコードって、表の行を指してるからね。)
