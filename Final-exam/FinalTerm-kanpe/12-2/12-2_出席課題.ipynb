<img src="images/image1.png" style="height:400px;background-color:white">

1. 上記の図に示される分散システム構成において、単一障害点となるコンピュータノードをすべて列挙せよ


        1台しかないコンピュータノードの種類を選べば正解

        - ファイルサーバ (File Server)
        - リバースプロキシ (Reverse Proxy)
        - データベースサーバ (Database Server)  

        (なお、今回は「コンピュータノード」なので対象外ですが、場合によってはネットワーク)
        回線などコンピュータノード以外が単一障害点となる可能性もあります。

2. 単一障害点となっているノードの1つがダウンした場合と、そうでないノードが1つダウンした場合にどのような結果となるか、違いがわかるように説明せよ

        単一障害点となっているノードがダウンした場合、分散システムが提供している
        サービスが停止状態あるいは機能不全となった状態となってしまう。
        一方で、単一障害点ではないノードがダウンした場合は、パフォーマンスの劣化など
        が発生する可能性はあるが、分散システムが提供するサービスを完全な状態で
        維持することができる。
