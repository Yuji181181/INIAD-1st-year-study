## Webアプリケーション開発

***

### UMLとは
統一モデリング言語(Unified Modeling Language)
- システムやアプリケーション開発時の設計図として利用されている
- UML2.0以降でダイアグラムが13種類ある

ダイアグラムの表記による分類は以下の通り：

![](images/ダイアグラム.png)



### UML2.0のダイアグラムの中から3つ紹介：
#### 1. ユースケース図

![](images/ユースケース図.png)

例：「大学の講義」
- 講義にかかわる「アクター」
    - 学生
    - 事務員
    - 教員
- 講義の提供する「ユースケース」
    - 「履修登録をする」：学生が履修登録をし、事務が確認する
    - 「講義に出席する」：学生が講義に出席し、教員が出欠を確認する 
    - 「試験を受ける」：学生が試験を受け、教員が採点する
    - 「成績登録をする」：教員が成績登録を行い、事務が確認する

例：「ショッピングサイト」
- 「アクター」
    - ユーザ
    - 出店者
- 「ユースケース」
    - 「商品を登録する」：出店者が商品を登録する
    - 「商品を検索する」：ユーザが登録された商品を検索する
    - 「商品を注文する」：ユーザが商品を注文し、出店者が受ける
    - 「注文をキャンセルする」：ユーザが注文をキャンセルし、出店者が処理をする

#### 2. クラス図

![](images/クラス図.png)

- クラス図の関係の記法は以下の通り：

![](images/関係の記法.png)

例：「自動車」
- 集約関係
    - 1つの「自動車」には4つの「タイヤ」がある
    - 1つの「自動車」には1つの「エンジン」がある
- 汎化関係
    - 「バス」は「自動車」の1種である
    - 「タクシー」は「自動車」の1種である
- 関連（特別な意味のない関係）
    - 「自動車メーカー」は複数の「自動車」を作っている

例：「ブログ」
- 次のようなブログシステムを考える
    - 「ユーザ」は「記事」を投稿することができる
    - 「ユーザ」は「記事」に「コメント」をつけることができる
- 右のようなクラス図でモデル化できる
    - 「ユーザ」は複数の「記事」を持つ
    - 「記事」は複数の「コメント」を持つ
    - 「ユーザ」は複数の「コメント」を持つ

#### 3. シーケンス図

![](images/シーケンス図.png)

例：レストラン
- 「オブジェクト」
    - 客
    - ウェイター
    - コック
- 「メッセージ」
    - 客がウェイターに「飲み物をオーダー」
      → ウェイターが客に「飲み物をサーブ」
    - 客がウェイターに「料理をオーダー」
      → 調理が完了した後に、ウェイターが客に「料理をサーブ」
    - ウェイターがコックに「調理を依頼」
      → コックがウェイターに「調理完了」の連絡
