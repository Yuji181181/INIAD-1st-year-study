#### 出席課題

1. あるDjangoプロジェクトに、newsというアプリケーションがあり、以下に示すようなURLパターンが定義されている。

> from django.contrib import admin
> from django.urls import path
> from news import views
> 
> urlpatterns = [
>     path('admin/', admin.site.urls),
>
>     path('news/', views.news_list, name='news_list'),
>
>     path('news/<int:news_id>/', [   A   ], name='news_detail'),
>
>     path('news/search',views.news_search, name='news_search'),
>
>     path('[   B   ]', views.news_create, name='news_create'),
>
> ]

DjangoのWebサーバを動かした時、以下のような動作をした。

- http://127.0.0.1:8000/news : views.news_list が呼ばれる
- http://127.0.0.1:8000/news/10/ : views.news_detail が呼ばれる
- http://127.0.0.1:8000/news/search : views.news_search が呼ばれる
- http://127.0.0.1:8000/news/create : views.news_create が呼ばれる

この時、URLパターン内の空白を埋めよ。

> A : views.news_detail
>
> B : news/create

2. このnewsアプリケーションでは、下に示すようなModelが定義されているとする。

> from django.db import models
> 
> class Content(models.Model):
>     title = models.CharField(max_length=50)
>
>     text = models.TextField()
>
>     posted_at = models.DateTimeField('date published')

以下のURLをWebブラウザで開いた際には、全てのニュースのタイトルと本文を一覧表示したい。

- http://127.0.0.1:8000/news

この時、news_list 関数およびテンプレートファイル news_list.html の空白を埋めよ。
