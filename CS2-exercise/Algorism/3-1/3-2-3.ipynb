```python
import time
import random
import matplotlib.pyplot as plt

# 選択ソートの関数
def selection_sort(lst):
    length = len(lst)
    for i in range(length):
        minpos = i
        for j in range(i + 1, length):
            if lst[j] < lst[minpos]:
                minpos = j
        if i != minpos:
            tmp = lst[i]
            lst[i] = lst[minpos]
            lst[minpos] = tmp

# リストサイズごとの計算時間を記録
sizes = list(range(100, 1100, 100))
times = []

for size in sizes:
    test_list = [random.randint(0, 10000) for _ in range(size)]
    start_time = time.time()
    selection_sort(test_list)
    end_time = time.time()
    times.append((end_time - start_time) * 1000)  # ミリ秒に変換

# プロット
plt.plot(sizes, times, 'o-')
plt.xlabel('n')
plt.ylabel('time (ms)')
plt.title('Time of selection_sort(n)')
plt.show()
```
