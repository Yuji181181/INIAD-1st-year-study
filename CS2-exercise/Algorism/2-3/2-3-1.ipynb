# 課題１
以下の５つの関数について、1≦x≦8の区間のグラフを重ねて描画する

- $y = x$
- $y = x^2$
- $y = 2^x$
- $y = logx$
- $y = xlogx$

```python
import matplotlib.pyplot as plt
import math
```

```python
import math
import matplotlib.pyplot as plt

x_values = []
y1_values = []
y2_values = []
y3_values = []
y4_values = []
y5_values = []

x = 1
while x <= 8:
    x_values.append(x)
    y1_values.append(x)        
    y2_values.append(x ** 2) 
    y3_values.append(2**x)   
    y4_values.append(math.log(x)) 
    y5_values.append(x * math.log(x)) 
    x += 0.1

plt.plot(x_values, y1_values, label="$y = x$", color='blue')  
plt.plot(x_values, y2_values, label="$y = x^2$", color='orange') 
plt.plot(x_values, y3_values, label="$y = 2^x$", color='green') 
plt.plot(x_values, y4_values, label="$y = log(x)$", color="red")
plt.plot(x_values, y4_values, label="$y = x * log(x)$", color='purple')

plt.title('Graphs of several functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
x_values = []
y_values = []

for x in range(-100, 100): 
    x_float = x / 10.0
    if x_float != 0:
        x_values.append(x_float)
        y_values.append(1/x_float)

plt.plot(x_values, y_values, label='y = 1/x')

plt.title('Graph of y = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.xlim(-10, 10) 
plt.legend()
```
