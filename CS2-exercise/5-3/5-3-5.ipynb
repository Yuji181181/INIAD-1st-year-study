```python
from collections import deque
import random

attraction = [deque(), deque(), deque()]
next_pool = deque()
goal = deque()

class Visitor:
    def __init__(self, name, lst):
        self.name = name
        self.attr = deque(lst)

# 客の数をNに設定
N = 5  # 例として5人に設定していますが、任意の数に変更可能です

# 客のリストを作成
visitors = []
for i in range(N):
    name = chr(65 + i)  # A, B, C, ... と名前をつける
    attractions = list(range(3))  # 0, 1, 2 のアトラクション順序
    random.shuffle(attractions)  # ランダムな順序にする
    visitors.append(Visitor(name, attractions))

# 客の情報を表示
for visitor in visitors:
    print(f"{visitor.name}さんのアトラクション訪問順序", visitor.attr)
print("")

num_of_visitors = N

# 客をnext_poolに追加
next_pool.extend(visitors)

def next_action(visitor):
    if len(visitor.attr) > 0:
        i = visitor.attr.popleft()
        print(f"{visitor.name}さんはアトラクション{i}に並びました。")
        attraction[i].append(visitor)
    else:
        print(f"{visitor.name}さんはアトラクションを回り終えました。")
        goal.append(visitor)

for t in range(100):
    print("t=", t)

    for i in range(len(attraction)):
        if len(attraction[i]) > 0:
            v = attraction[i].popleft()
            next_pool.append(v)
    
    while len(next_pool) > 0:
        v = next_pool.popleft()
        next_action(v)
    
    print("")
    if len(goal) == num_of_visitors:
        print("終了時刻", t)
        break
```
