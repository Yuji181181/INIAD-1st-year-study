```python
months = []
temperatures = []
data = [
    "1,11.1,3.7,7.1,135.0",
    "2,13.3,4.0,8.3,15.0",
    "3,16.0,6.2,10.7,131.0",
    "4,18.2,7.9,12.8,296.5",
    "5,24.0,15.6,19.5,118.0",
    "6,27.5,19.8,23.2,212.5",
    "7,27.7,21.8,24.3,270.5",
    "8,34.1,25.3,29.1,61.5",
    "9,28.1,21.5,24.2,117.5",
    "10,21.4,14.4,17.5,205.0",
    "11,18.6,10.1,14.0,14.5",
    "12,12.3,3.7,7.7,13.0"
]

for line in data:
    parts = line.split(',')
    months.append(int(parts[0]))
    temperatures.append(float(parts[1]))

plt.plot(months, temperatures, label='Temp.', marker='o')
plt.title('Tokyo monthly mean temperature in 2020')
plt.xlabel('Month')
plt.ylabel('°C')
plt.xticks(months) 
plt.legend()
```
