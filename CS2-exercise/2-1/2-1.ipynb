```python
import matplotlib.pyplot as plt
import random
import time

lengths = [100, 500, 1000, 5000, 10000]
linear_times = []
binary_times = []

def linear_search(arr, target):
    for i in arr:
        if i == target:
            return True
    return False

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return True
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return False

for length in lengths:
    data = [random.randint(0, 1000000) for _ in range(length)]
    target = random.randint(0, 1000000)

    start_time = time.time()
    for _ in range(10000):
        linear_search(data, target)
    linear_times.append((time.time() - start_time) * 1000)  # ms

    data.sort()
    
    start_time = time.time()
    for _ in range(10000):
        binary_search(data, target)
    binary_times.append((time.time() - start_time) * 1000)  # ms

plt.plot(lengths, linear_times, 'o-', label='Linear search')
plt.plot(lengths, binary_times, 'x-', label='Binary search')
plt.xlabel('Length')
plt.ylabel('Time (ms)')
plt.title('Linear search vs Binary search')
plt.legend()
```
